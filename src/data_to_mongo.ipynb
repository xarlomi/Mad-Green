```python
def porcentaje (item):
    '''
    Esta funcion nos divide la col principal dee producto en dos: 
    nombre del producto y % de cbd del product
    '''
    item = item.split(" ")
    for i in item:
        if "%" in i:
            numero = i
    
        
    return numero
```

```python
scrapeo['CBD percentage'] = scrapeo['Product'].apply(porcentaje)
```
